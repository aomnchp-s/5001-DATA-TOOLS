```python
import pandas as pd
import numpy as np
df = pd.read_csv('2021-2022.csv')
```

# ข้อ1:  5 อันดับทีมที่ได้ใบแดงมากที่สุด

```python
redHome_df = df[['HomeTeam','HR']]
result_redHome = redHome_df.groupby('HomeTeam')['HR'].sum().reset_index()

redAway_df = df[['AwayTeam','AR']]
result_redAway = redAway_df.groupby('AwayTeam')['AR'].sum().reset_index()

red_card =result_redHome.merge(result_redAway, left_on='HomeTeam', right_on='AwayTeam')
red_card['Total_Red'] = red_card['HR']+red_card['AR']
redcards = red_card[['HomeTeam','Total_Red']].sort_values(by=['Total_Red'],ascending=False).set_axis(['Team','Total Red Card'], axis='columns')
redcards.head(5)
```

# ข้อ2: 5 อันดับทีม ที่ทำประตูได้มากที่สุด

```python
hometeam_df = df[['HomeTeam','FTHG']]
result1 = hometeam_df.groupby('HomeTeam')['FTHG'].sum().reset_index(name='Count_FTHG')

AwayTeam_df = df[['AwayTeam','FTAG']]
result2 = AwayTeam_df.groupby('AwayTeam')['FTAG'].sum().reset_index(name='Count_FTAG')

team_goals =result1.merge(result2, left_on='HomeTeam', right_on='AwayTeam')
#team_goals
team_goals['Total_Goals'] = team_goals['Count_FTHG']+team_goals['Count_FTAG']
maxdf = team_goals[['HomeTeam','Total_Goals']].sort_values(by=['Total_Goals'],ascending=False).set_axis(['Team','Total Goals'], axis='columns')
maxdf.head(5)
```

# ข้อ3: เปอร์เซ็นที่ทำคะแนนประตูได้จากการยิงประตูของแต่ละทีม (ทำคะแนนประตูได้คิดเป็นกี่เปอร์จากการยิงประตู)

```python
home_shots_df = df[['HomeTeam','HS','FTHG']]
result_shots1 = home_shots_df.groupby('HomeTeam')['HS','FTHG'].sum().reset_index()

away_shots_df = df[['AwayTeam','AS','FTAG']]
result_shots2 = away_shots_df.groupby('AwayTeam')['AS','FTAG'].sum().reset_index()

team_shots =result_shots1.merge(result_shots2, left_on='HomeTeam', right_on='AwayTeam')
#team_shots

team_shots['Total_Shots'] = team_shots['HS']+team_shots['AS']
team_shots['Total_Goals'] = team_shots['FTHG']+team_shots['FTAG']
#team_shots

team_shots['Percent'] = (team_shots['Total_Goals']*100)/team_shots['Total_Shots']
team_shots
team_shots[['HomeTeam','Total_Shots','Total_Goals','Percent']].sort_values(by=['Percent'],ascending=False).round(2).set_axis(['Team','Total Shots','Total Goals','Percent (%)'], axis='columns')
```

# ข้อ4: เปรียบเทียบเปอร์เซ็นชนะการแข่งขันระหว่างเป็นทีมเจ้าบ้านและทีมเยือน

```python
hometeam_win = df[['HomeTeam','FTR']]
homewin = hometeam_win.groupby(['HomeTeam','FTR'])['FTR'].count().reset_index(name='H_times')
homewin2 = homewin.groupby('HomeTeam')['H_times'].sum().reset_index(name='total_times')

totalHomeWin = homewin.merge(homewin2, left_on='HomeTeam', right_on='HomeTeam')
totalHomeWin['Home Percent'] = (totalHomeWin['H_times']*100)/totalHomeWin['total_times']
homedf = totalHomeWin[['HomeTeam','FTR','Home Percent']]
homedf2 = homedf[homedf.FTR=='H']
#xx

awayteam_win = df[['AwayTeam','FTR']]
awaywin = awayteam_win.groupby(['AwayTeam','FTR'])['FTR'].count().reset_index(name='A_times')
awaywin2 = awaywin.groupby('AwayTeam')['A_times'].sum().reset_index(name='total_times')

totalAwayWin = awaywin.merge(awaywin2, left_on='AwayTeam', right_on='AwayTeam')
totalAwayWin['Away Percent'] = (totalAwayWin['A_times']*100)/totalAwayWin['total_times']
awaydf = totalAwayWin[['AwayTeam','FTR','Away Percent']]
awaydf2 = awaydf[awaydf.FTR=='A']
#xx2

df_merge = homedf2.merge(awaydf2, left_on='HomeTeam', right_on='AwayTeam')
df_merge[['HomeTeam','Home Percent','Away Percent']].round(2).set_axis(['Team','Home Percent (%)','Away Percent (%)'], axis='columns')
```
