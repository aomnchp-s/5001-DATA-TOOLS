```python
import pandas as pd
df = pd.read_csv('2021-2022.csv')
```

# ข้อ1: ทีมที่ทำประตูได้มากที่สุดและน้อยที่สุดใน EPL2021~2022 (เรียงจากมาก-น้อย)

```python
hometeam_df = df[['HomeTeam','FTHG']]
result1 = hometeam_df.groupby('HomeTeam')['FTHG'].sum().reset_index(name='Count_FTHG')

AwayTeam_df = df[['AwayTeam','FTAG']]
result2 = AwayTeam_df.groupby('AwayTeam')['FTAG'].sum().reset_index(name='Count_FTAG')

team_goals =result1.merge(result2, left_on='HomeTeam', right_on='AwayTeam')
#team_goals
team_goals['Total_Goals'] = team_goals['Count_FTHG']+team_goals['Count_FTAG']
team_goals[['HomeTeam','Total_Goals']].sort_values(by=['Total_Goals'],ascending=False)
```

# ข้อ2:เปอร์เซ็นที่ทำประตูได้จากการยิงประตูของแต่ละทีม (ทำประตูได้คิดเป็นกี่เปอร์จากการยิงประตู)

```python
home_shots_df = df[['HomeTeam','HS','FTHG']]
result_shots1 = home_shots_df.groupby('HomeTeam')['HS','FTHG'].sum().reset_index()

away_shots_df = df[['AwayTeam','AS','FTAG']]
result_shots2 = away_shots_df.groupby('AwayTeam')['AS','FTAG'].sum().reset_index()

team_shots =result_shots1.merge(result_shots2, left_on='HomeTeam', right_on='AwayTeam')
#team_shots

team_shots['Total_Shots'] = team_shots['HS']+team_shots['AS']
team_shots['Total_Goals'] = team_shots['FTHG']+team_shots['FTAG']
#team_shots

team_shots['Percent'] = (team_shots['Total_Goals']*100)/team_shots['Total_Shots']
team_shots
team_shots[['HomeTeam','Total_Shots','Total_Goals','Percent']].sort_values(by=['Percent'],ascending=False).round(2)
```

```python
#ทีมเจ้าบ้านและทีมเยือน ที่แข่งชนะมากที่สุดและน้อยที่สุด
df_HomeTeam = df[['HomeTeam','FTR']].where(df.FTR=='H')
result_HomeTeam = df_HomeTeam.groupby('HomeTeam')['FTR'].size().reset_index(name='Count')

df_AwayTeam = df[['AwayTeam','FTR']].where(df.FTR=='A')
result_AwayTeam = df_AwayTeam.groupby('AwayTeam')['FTR'].size().reset_index(name='Count')

print('ทีมเจ้าบ้าน แข่งชนะมากที่สุด')
print(result_HomeTeam[result_HomeTeam.Count == result_HomeTeam.Count.max()])
print()
print('ทีมเจ้าบ้าน แข่งชนะน้อยที่สุด')
print(result_HomeTeam[result_HomeTeam.Count == result_HomeTeam.Count.min()])
print()

print('ทีมเยือน แข่งชนะมากที่สุด')
print(result_AwayTeam[result_AwayTeam.Count == result_AwayTeam.Count.max()])
print()
print('ทีมเยือน แข่งชนะน้อยที่สุด')
print(result_AwayTeam[result_AwayTeam.Count == result_AwayTeam.Count.min()])
```
